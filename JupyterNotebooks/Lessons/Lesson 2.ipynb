## Basic HTML

This section is going to be mostly informational **BUT** will be used a lot for the front end portion of things!

_**Do we remember what HTML uses to write it's content?**_

This section is going to be talking about the various _______ that we can use to get started!

### \<!DOCTYPE html> DOCTYPE Tag
This helps us define the document type. This tells the browser what type of content is coming in the file.

### \<!-- This is a comment --> Comment Tag
Title says it all \<!-- --> is used to have comments in your HTML!

### \<html> HTML Tag
This defines the root of the HTML document. This will typically wrap everything BUT the DOCTYPE tag.

### \<head> Header Tag
This section contains all of our metadata for our webpage. The files that are linked to it. Title for the pages. Language the page is being served in etc.

### \<link> Link Tag
This tag defines the relationship between this document and an external one. Think linking stylesheets to your webpage. This is the most common application.

### \<meta> Metadata Tag
This tag will help you define metadata to the page. Things like the character set that you're using. Devices you support.

### \<script> Script Tag
Our script tags are used to attach our javascript logic to our webpage. We won't use this too much early on but this is where our webpages will get a lot of their magic from. 

### \<title> Title Tag
Allows us to give the webpage a title. This typically shows up in the tab so you can see what webpage you're on. ex. Youtube and Twitch below

![Tab Titles](../../Assets/Images/TabTitles.png)

### \<body> Body Tag
Now we can get to the meat and potatoes of our page. What goes in here is all of the content that we want our page to display. This is a wrapping tag like HTML and head rather than purely being a content tag.

### \<h1>, \<h2>, \<h3>, \<h4>, \<h5>, \<h6> Header Tags
These tags we use to catch attention. These are header tags, we have 6 of them. The numbers correspond to the size of the header. 1 being the largest and 6 being the smallest.

### \<p> Paragraph Tag
This is a paragraph tag. We use this for large blocks of text. This is one that will typically be used often. 

### \<div> Division Tag
These are probably my personal favorite and also at one time were my most hated. Division tags are used to split the page up into sections. When learning how to use them life can be a little frustrating. Once you get a grip on them they're an amazing tool to have!

### \<a> Anchor Tag
Why it's called the Anchor tag, I don't really know. What I do know is that this is the tag that we use when we want links on our page to another site or other pages on our own sites!

### \<img> Image Tag
Simple! It's how we add images to our web page!

### \<br> Break Tag
We can use this to add a line break in our webpage. Think new line in HTML form.

### \<footer> Footer Tag
I personally use this to section off the bottom part of a webpage. It isn't necessary though.

### \<form> Form Tag
We're going to need these further down the road. Form tags help us submit data to our back end. This is going to be a large component of how your users will submit data on your web pages. 

### \<ul> Unordered List Tag
This is a wrapper tag to create an unordered list

### \<ol> Ordered List Tag
This is a wrapper tag to create an ordered list

### \<li> List Tag
This will go inside of our ul/ol list elements to be data in the list

### \<table> Table Tag
This is a wrapper tag to create a table

### \<th> Table Header Tag
This is to define the header title at the top of the table row

### \<tr> Table Row Tag
This is a wrapper tag to create a row inside the table

### \<td> Table Data Tag
This tag is used to add data to the table


Now I know that this was a lot and it's far from the end on tags. This was a great place to start and hopefully a good resource/cheat sheet to look back at. There are a lot of tags that are super niche. I tried to avoid those for the time being.

If you'd like to look more into the tags. [Here is an HTML Element Reference](https://www.w3schools.com/TAGS/default.ASP)

## Getting started with Stylesheets

Now that we've talked a little about our skeleton with our HTML. We need to have the ability to add some of our style elements to it as well. 

We're going to stick to the basics for now. We have 3 primary ways that we can manipulate our HTML. By tag, by class, and by id. These are all done the same way and use the same attributes but are accessed differently.

### Selectors on Elements
First up are our tags. These can also be called elements and they typically are in CSS and we usually use them like this:

```
p {
    font-size: 12px;
}
```

Putting style on elements is typically not done a lot.

_**Can we think of any reasons why?**_

### Selectors on Classes
Next we have classes. Classes are very commonly used to group similar elements. We can have multiple classes on a single element, meaning that we can stack properties on an element. This does require you to get a grasp on how certain properties interact and mesh with one another.

Classes are typically denominated with a `.` so in our CSS it would look like this:
```
.white {
    color: white;
}
```

### Selectors on IDs
Finally we have IDs. This is the mose selective of the three. IDs should ONLY be assigned to one element. Ignoring this can cause your webpage to act in unpredictable ways. 

Other than that we act pretty similarly to what's come before with the properties and the syntax. Instead of `.` we use `#`. So denomination for IDs would look like this:

```
#firstParagraph {
    margin: auto 1.6%;
    width: 30%;

}
```

## Agile Roles and Getting Started with an Agile Project

Being that you're in college and not working a full time job we're not going to be excucuting agile to a tee. Think is more so to get you some exposure to what working in an agile manner would be like. 

So we're going to talk about the roles that correspond with agile. 

First thing to remember before we go into the roles is that I am your customer! 

### Developers
So this one is pretty self explainitory but at the same time it isn't. So yes you do have developers in the squad. What agile did with the term is expand it to apply to anyone on the development team. You're all intregal to developing the product, so you're all developers. Developers can include your IT Specialist, Designers, Doc Writers, Support Team Members, etc. 

Developers and the development team should be self managing and be able to assign and handle managing themselves. The idea is that the folks on the team that wrote the code are the most likely to be able to implement changes and make fixes so they should have the ability to do so without major interference from above. 

You should be working on accomplishing the goals for the sprint and participating in daily standups to share status with your team.

### Product Owner
You are the person that everyone will talk to about about the plans for the product. You need to know what direction that the development team is headed in and make sure that they're aligned with the customers vision as well. It's also your responsibility to be in communication with the customer in regards to what they want to be done. You essentially need to be a planning guru and be able to manage multiple communication channels. 

Something that you are not though is a dictator! You're responsibility is to keep everyone on the same page. NOT to tell everyone what to do. Remember the squad is self managing!

You're responsible for customer or stakeholder management, managing the Scrum Backlog, and Release management.

### Scrum Master/Iteration Manager
You're responsible for the other half of interacting with the team. With the PO we talked about communication with the customer. For the Scrum Master/Iteration Manager, you're responsible for managing the communication within the team itself. This means running the standups, holding iteration retrospectives, keeping transparancy within the team, helping avoid blockers, among other things!

With this you do play a large role daily and are often looked at to keep things running smoothly internally to the team. This is a very communicative role as well!

While you are responsible for keeping communication up you shouldn't be taking on others responsibilities. If someone isn't chipping in or doing their part this is still a management (Professor) issue!

[For more information on the roles look here](https://www.atlassian.com/agile/scrum/roles)
